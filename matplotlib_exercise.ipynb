```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#
#Follow along with these steps: **
#Create a figure object called fig using plt.figure() **
#Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
#Plot (x,y) on that axes and set the labels and titles to match the plot below:**

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Label')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.plot(x,y)
```

```python
#** Create a figure object and put two axes on it, ax1 and ax2. 
# Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

#** Now plot (x,y) on both axes. And call your figure object to show it.**

ax1.plot(x,y,color='red')
ax2.plot(x,y,color='red')


```

```python
#** Create the plot below by adding two axes to a figure object 
# at [0,0,1,1] and [0.2,0.5,.4,.4]**

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

#** Now use x,y, and z arrays to recreate the plot below. 
# Notice the xlimits and y limits on the inserted plot:**

ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax1.plot(x,z)

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
ax2.plot(x,y)
```

```python
#use plt.subplots(nrows=1, ncols=2) to create the plot below.**

plt.subplot(1,2,1)
plt.plot(x,y,'r',lw=2,linestyle='-.')

plt.subplot(1,2,2)
plt.plot(x,z,'b',lw=4)

#** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style** bv
```

```python
# figure size and aspect ratio
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(x,z)
plt.tight_layout()
```
