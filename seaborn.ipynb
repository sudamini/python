```python
#seaborn is a data visualization library
import seaborn as sns
```

```python
#see your visualizations inside of a notebook
%matplotlib inline
```

```python
#seaborn
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
#dist plot allows us to see a distribution for univariate set of observations
sns.distplot(tips['total_bill'])
#you will get a histogram and KDE = Kernel Density Estimation which is the line.
```

```python
#you can remove KDE 
sns.distplot(tips['total_bill'], kde=False)
```

```python
#give a third parameter as bins to get more information
sns.distplot(tips['total_bill'], kde=False, bins=30)
```

```python
#if you put a number thats too high for bins, then it starts to plot every single data point and then your data doesnt 
#look clear
sns.distplot(tips['total_bill'], kde=False, bins=100)
```

```python
#joint plot allows you to plot two distplots by byvariate data.
#you can essentially combine two different distplots.
#jointplot is usually used with default scatter because that is the easiest to read.
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#the hexagon that has more number of data points will be darker in color.
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
#kind=reg shows a linear fit to a scatter point data
sns.jointplot(x=['total_bill'],y=['tip'],data=tips, kind='reg')
```

```python
#kind=kde shows a two dimensional density plot
sns.jointplot(x=['total_bill'],y=['tip'],data=tips,kind='kde')
```

```python
#pairplot plots the pairwise relationship across entire dataset ( for numerical data points )
#pairplot will do a jointplot for every single combination of numerical columns in our dataset
#larger your dataframe, longer it takes to plot your data.
sns.pairplot(tips)
```

```python
#you can pass an argument 'hue' to pairplot.
#hue means a categorical column which has two categories e.g. male and female.
sns.pairplot(tips,hue='sex')
```

```python
#pallete can choose color palette
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#it draws a dash for each data point in your dataset
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
#categorical plots
#distribution of categorical columns wrt some numerical column or another categorical column
sns.barplot(x='sex',y='total_bill',data=tips)
#this shows the average total bill against the categorical column sex
#average total bill for males is slightly higher than average total bill for females.
```

```python
#you can add another argument to it called estimator
#estimator is some sort of a statistical function e.g. standard deviation
import numpy as np
#np.std function for standard deviation
#x needs to be categorical column
#y needs to be numerical column
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#countplot is similar to a barplot except the estimator is counting the number of occurrences.
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
#box plot is used to show distribution of categorical data
#it shows distribution of quantitative data to facilitate comparison between variables
sns.boxplot(x='day',y='total_bill',data=tips)
#quartiles show distribution of datasets
#points above the whiskers show outliers
```

```python
#we can get really fancy with seaborn by adding some more options
#hue = split up these box plots by more categorical columns
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
#e.g. on fridays, smokers tend have a larger bill than non-smokers
```

```python
#violinplot is similar to a boxplot
sns.violinplot(x='day',y='total_bill',data=tips)
#x will be categorical column
#y will be numerical column
#violinplot allows us to plot all the components across all datapoints
#it shows density distribution of underlying distribution
```

```python
#violinplots have a lot more information and are harder to read.
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
#you can split violinplot so that two sexes are next to each other
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#stripplot is a scatter plot based off of category
sns.stripplot(x='day',y='total_bill',data=tips)
#in this plot, you cannot tell how many points are stacked on top of each other
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
#jitter=True adds a little bit of random noise to separate some of these stacked points.
```

```python
#you can add hue and split
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#swarmplot combines stripplot,violinplot and scatterplot
sns.swarmplot(x='day',y='total_bill',data=tips)
#they dont scale too well for very large data sets
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
#these plots are more useful for explanatory data analysis
```

```python
#factorplot is a general form of all these plots
sns.catplot(x='day',y='total_bill',data=tips, kind='bar')
```
